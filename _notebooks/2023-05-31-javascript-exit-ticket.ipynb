```python
%%js
console.log("Examine Data Types");

// Function to add typeof to output
function getType(output) {
    return typeof output + ": " + output;
}

// Function defintion
function logIt(output) {
    console.log(getType(output));  // logs string
    console.info(output);          // logs object
    element.append(getType(output) + "<br>");  // adds to Jupyter output
    //alert(getType(output));
}

// Common Types
element.append("Common Types <br>");
logIt("Mr M"); // String
logIt(1997);    // Number
logIt(true);    // Boolean
element.append("<br>");

// Object Type, this definition is often called a array or list
element.append("Object Type, array <br>");
var scores = [
    90,
    80, 
    100
];  
logIt(scores);
element.append("<br>");

// Complex Object, this definition is often called hash, map, hashmap, or dictionary
element.append("Object Type, hash or dictionary <br>");
var person = { // key:value pairs seperated by comma
    "name": "Mr M", 
    "role": "Teacher"
}; 
logIt(person);
logIt(JSON.stringify(person));  //method used to convert this object into readable format
```

```python
%%js
console.log("Classroom object");

class mapImage {
  /* method: constructor
   * parameters: name, src (url for the map's image)
   * description: returns object when "new mapImage()" is called with matching parameters
   * assignment: this.name, this.src are properties retained in the returned object
  */
  constructor(name, src, role) {
    this.name = name;
    this.src = src;
    this.role = role;
  }

  /* method: setter
   * parameters: role - role in classroom
   * description: this.role is updated from default value to value contained in role parameter
  */
  setRole(role) {
    this.role = role;
  }
                            
  function displayEasy() {
    // define a easy Array of mapImage objects
    const easy = [
        new mapImage("Basketball Courts", "images/dn1.png", "Easy"),
        new mapImage("Gym", "images/dn2.png", "Easy"),
        new mapImage("Pool", "images/dn3.png", "Easy"),
        new mapImage("Stadium Courts", "images/dn4.png", "Easy"),
        new mapImage("Tennis Courts", "images/dn5.png", "Easy")
    ];
    return new Difficulty(easy);  // returns object
  }
  
  /* method: getter
   * description: turns properties of object into JSON object
   * return value: JSON object
  */
  getJSON() {
    const obj = {type: typeof this, name: this.name, src: this.src, role: this.role};
    const json = JSON.stringify(obj);
    return json;
  }

  /* method: logIT
   * description: "this" mapImage object is logged to console
  */
  logIt() {
    //mapImage Object
    console.info(this);
    //Log to Jupter
    element.append("mapImage json <br>");
    element.append(this.getJSON() + "<br>");  
    //alert(this.getJSON());
  }
    
}

/* class: Difficulty
 * Description: A collection of mapImage objects
*/
class Difficulty {
  /* method: constructor
   * parameters: easy, medium, and hard - an array of mapImage objects
   * description: returns object when "new Difficulty()" is called containing properties and methods of the difficulties of levels
   * assignment: this.difficulty, this.easy, ... are properties retained in the returned object
  */
  constructor(easy, medium, hard) {
    /* spread: this.difficulty contains all mapImage objects
     * map: this.json contains of map of all persons to JSON
    */
    this.easy = easy;
    this.medium = medium;
    this.hard = hard;
    this.difficulty = [...easy, ...medium, ...hard]; // ... spread option
    this.json = '{"difficulty":[' + this.difficulty.map(mapimage => mapimage.getJSON()) + ']}';
  }

  /* method: logIT
   * description: "this" Difficulty object is logged to console
  */
  logIt() {
    //Difficulty object
    console.log(this);
    
    // json
    element.append("Difficulty object in JSON<br>");
    element.append(this.json + "<br>");  
    //alert(this.json);
  }
}

/* function: constructCompSciClassroom
 * Description: Create data for Classroom and Person objects
 * Returns: A Classroom Object
*/
function displayImages() {
    // define a easy Array of mapImage objects
    const easy = [
        new mapImage("Basketball Courts", "images/dn1.png", "Easy"),
        new mapImage("Gym", "images/dn2.png", "Easy"),
        new mapImage("Pool", "images/dn3.png", "Easy"),
        new mapImage("Stadium Courts", "images/dn4.png", "Easy"),
        new mapImage("Tennis Courts", "images/dn5.png", "Easy")
    ];
    
    // define a medium Array of mapImage objects
    const medium = [
        new mapImage("Cafeteria", "images/dn6.png", "Medium"),
        new mapImage("Library", "images/dn7.png", "Medium"),
        new mapImage("Quad", "images/dn8.png", "Medium"),
        new mapImage("Office", "images/dn9.png", "Medium"),
        new mapImage("Locker Room", "images/dn10.png", "Medium"),
        new mapImage("PAC", "images/dn11.png", "Medium")
    ];

    // define a hard Array of mapImage objects
    const hard = [
        new mapImage("A Building", "images/dn12.png", "Hard"),
        new mapImage("D Building", "images/dn13.png", "Hard"),
        new mapImage("E Building", "images/dn14.png", "Hard"),
        new mapImage("G Building", "images/dn15.png", "Hard"),
        new mapImage("J Building", "images/dn16.png", "Hard"),
        new mapImage("K Building", "images/dn17.png", "Hard"),
        new mapImage("L Building", "images/dn18.png", "Hard"),
        new mapImage("P Building", "images/dn19.png", "Hard")
    ];

    // make a CompSci classroom from formerly defined teacher and student objects
    return new Difficulty(easy, medium, hard);  // returns object
}

// assigns "compsci" to the object returned by "constructCompSciClassroom()" function
const images = displayImages();
// output of Objects and JSON in CompSci classroom
images.logIt();
// enable sharing of data across jupyter cells
$('#jsonText').text(images.json);  // posts/embeds/writes compsci.json to HTML DOM element called jsonText
                            
const eimages = displayEasy();
// output of Objects and JSON in CompSci classroom
eimages.logIt();
// enable sharing of data across jupyter cells
$('#jsonText').text(eimages.json);  // posts/embeds/writes compsci.json to HTML DOM element called jsonText
```

```python
%%html
<!-- load jQuery and tablesorter scripts -->
<html>
    <head>
        <!-- load jQuery and tablesorter scripts -->
        <script src="https://ajax.googleapis.com/ajax/libs/jquery/3.6.0/jquery.min.js"></script>
        <script src="https://cdnjs.cloudflare.com/ajax/libs/jquery.tablesorter/2.31.3/js/jquery.tablesorter.min.js"></script>
        <style>
            /* CSS-style selector maps to table id or other id's in HTML */
            #jsonTable, #flaskTable {
                background-color: #353b45;
                padding: 10px;
                border: 3px solid #ccc;
                box-shadow: 0.8em 0.4em 0.4em grey;
            }
        </style>
    </head>

    <body>
        <!-- Table for writing and extracting jsonText -->
        <table id="jsonTable">
            <thead>
                <tr>
                    <th>Difficulty JSON Data</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td id="jsonText">{"difficulty":[{"type":"object","name":"sample","src":"sample","role":"sample"}]}</td>
                </tr>
            </tbody>
        </table>
        <br>
        <br>
        <br>
        <div class="container">
        <!-- Formatting for the map image -->
        <div class="row justify-content-md-center">
            <div class="col-md-auto bg-secondary text-white p-3">
                <!-- Image that changes based on user selection -->
                <img src="images/dndefault.png" alt="Map image" id="mapImage" style="width: 344px; height: auto"/>
            </div>

            <!-- Formatting and style for title and description -->
            <div class="col col-lg-4 p-3 ">
                <div class="alert alert-primary" role="alert">
                    <h3>Del Norte GeoGuesser:</h3> <p>This game is designed to help new students learn the locations of Del Norte in a fun and easy to use manner! </p>
                </div>
                <!-- Selection of the game difficulty that calls to the changeDifficulty function with different parameters -->
                <div class="btn-group btn-block pb-3 pt-2" role="group" aria-label="Choosing Difficulty">
                    <button type="button" class="btn btn-secondary" id="easy" onclick="changeDifficulty(1)">Easy</button>
                    <button type="button" class="btn btn-secondary" id="medium" onclick="changeDifficulty(2)">Medium</button>
                    <button type="button" class="btn btn-secondary" id="hard" onclick="changeDifficulty(3)">Hard</button>
                </div>

                <!-- Display the time remaining in the game -->
                <p id="timeRemaining"></p>

                <!-- Defining the section of the code that will change to display the game over message -->
                <div id="gameover">

                <!-- Input for user to give their guesses for the location -->
                <label for="inputAnswer">Guess the Location:</label><br>
                <input type="text" id="inputAnswer" placeholder="Type Here..." value="">
                <!-- Submit button that calls the function -->
                <button id="submitAnswer" onclick="getGuess()">Submit Answer</button>
                <br>
                <br>
                <!-- Display the result of the program -->
                <h6 class="border border-primary" id="endgamemessage"></h6>
                </div>

                <!-- Counter updates with the amount scored right and wrong -->
                <p id="countRight">Correct: 0</p>
                <p id="countWrong">Incorrect: 0</p>

            </div>
        </div>
    </div>
    
    </body>
</html>

<script>
%%js
console.log("Difficulty Web Page");

// extract JSON text from HTML page
const jsonText = document.getElementById("jsonText").innerHTML;
console.log(jsonText);
element.append("Raw jsonText element embedded in HTML<br>");
element.append(jsonText + "<br>");

// convert JSON text to Object
const difficulty = JSON.parse(jsonText).difficulty;
console.log(difficulty);

// from classroom object creates rows and columns in HTML table
element.append("<br>Formatted data sample from jsonText <br>");
for (var row of difficulty) {
    element.append(row.src + " " + row.name + '<br>');
    // tr for each row, a new line
    $('#difficulty').append('<tr>');
    // td for each column of data
    $('#difficulty').append('<td>' + row.name + '</td>');
    $('#difficulty').append('<td>' + row.src + '</td>');
    $('#difficulty').append('<td>' + row.role + '</td>');
    // tr to end row
    $('#difficulty').append('</tr>');
}

// Defining variables used throughout the program
let mapUrl;
let mapNumber;
let timeLeft;
let elem = document.getElementById('timeRemaining');
var timerId = setInterval(countdown, 1000);
let right = document.getElementById("countRight");
let wrong = document.getElementById("countWrong");
let button = document.getElementById("submitAnswer");
let countR = 0;
let countW = -1;

// Depending on the user's selection, there will be a different integer parameter defining the function
        function changeDifficulty(difficulty){
            // Setting the user input equal to the variable number
            let number = difficulty

            // Use of conditionals to determine which difficulty the user chose
            // Setting the variable mapUrl equal to the array that corresponding with each difficulty defined above
            // Setting the variable timeLeft equal to the amount of seconds that the user has to guess (50 seconds for easy, 40 seconds for medium, 20 seconds for hard)
            if (number === 1) {
                mapUrl = easy
                timeLeft = 50

            } else if (number === 2) {
                mapUrl = medium
                timeLeft = 40

            } else if (number === 3){
                mapUrl = hard
                timeLeft = 20
            }
            // Resetting the number of correct and incorrect guesses to zero
            right.innerHTML = "Correct: " + 0;
            wrong.innerHTML = "Incorrect: " + 0;
            // Selecting a random number from 0 to the length of the chosen map array and setting that equal to the variable mapNumber
            mapNumber = Math.floor(Math.random() * mapUrl.length);

            getrandomImg();
            getGuess();
        }

        // Changing the image src from the HTML to be the element with the index of random number from the chosen array
        function getrandomImg() {
            document.getElementById("mapImage").src = mapUrl[mapNumber];
            // Setting the count for right and wrong equal to zero
            countR = 0
            countW = 0
        }

        // Defining a new array
        mapNewlist = [];
        // Defining variable guess
        let guess;

        // Gathering the guess from the user and calling the function imgGame
        function getGuess(){
            guess = document.getElementById("inputAnswer").value;
            imgGame();
        }

        // Setting the guess equal to the variable guess2
        function imgGame(){
            let guess2 = guess;
            // Iterating through mapUrl array and removing the image location information (including the beginning "/static/assets/map/dn" and the ending ".png") and storing the result as a new array called mapNewlist
            for (let i = 0; i < mapUrl.length; i++) {
                mapNewlist[i] = mapUrl[i].slice(21,-4)
            }
            // Conditional determines whether or not the guess is equal to the name of the location
            // If correct, the winning text is displayed and 1 is added to the correct guesses counter
            if (guess2 == mapNewlist[mapNumber]) {
                document.getElementById("endgamemessage").innerHTML = "Congrats! The guess matches the location.";
                countR += 1;
                right.innerHTML = "Correct: " + countR;
            }
            // If the input is blank, no game text is displayed
            else if (guess2 == "") {
                document.getElementById("endgamemessage").innerHTML = null;
            }
            // If the input is incorrect, the losing text is displayed and 1 is added to the incorrect guesses counter
            else {
                document.getElementById("endgamemessage").innerHTML = "The guess doesn't match the location.";
                countW += 1;
                wrong.innerHTML = "Incorrect: " + countW;
            }
            // Setting the input back to zero to reset the game
            document.getElementById("inputAnswer").value = ""
        }
        // Timer counts down using conditional and replaces the div element with game over text once the timer is done
        function countdown() {
            if (timeLeft == -1) {
                clearTimeout(timerId);
                document.getElementById("gameover").innerText = "Game Over!"
            }

            else if (timeLeft == null){
                elem.innerHTML = ' ';
            }
            else {
                elem.innerHTML = timeLeft + ' seconds remaining';
                timeLeft--;
            }
        }

</script>
```
